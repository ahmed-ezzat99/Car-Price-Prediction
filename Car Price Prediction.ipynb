# Model Implementation

```python
X=data.iloc[:,1:] #features
Y=data.iloc[:,0]  #label
```

```python
#Feature Scaling
scaler = StandardScaler()
```

```python
#Splitting Dataset into Training and Testing
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=7)
```

```python
#Feature Scaling
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Implementing Random Forest Regressor and Evaluating Performance Metrics
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
model = RandomForestRegressor()
model.fit(X_train,Y_train)
```

# Accuracy

```python
#Assessing Performance Metrics of Random Forest Regression Model"
Y_pred = model.predict(X_test)
print(f"Mean Absolute Error: {mean_absolute_error(Y_test,Y_pred)}")
print(f"Mean Squared Error: {mean_squared_error(Y_test,Y_pred)}")
print(f"Root Mean Squared Error: {np.sqrt(mean_squared_error(Y_test,Y_pred))}")
print(f"R2 Score: {r2_score(Y_test,Y_pred)}")
```

```python
#Evaluating Model Accuracy
accuracy = ("The accuracy of model is : ", r2_score(Y_test, Y_pred)*100)
print(accuracy)
```

```python
#Scatter Plot of Actual and Predicted Values
plt.figure(figsize=(8,5))
plt.scatter(Y_test, Y_test, color='blue', label='Actual')
plt.scatter(Y_test, Y_pred, color='red', label='Predicted')

plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.legend()

plt.show()
```
